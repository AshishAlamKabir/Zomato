## **Importing Library**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
import sklearn
```

# **Read Files**

```python
df = pd.read_csv('/content/Zomato data .csv')
```

# **1. Find the most ordered restaurant type**

```python
df["rate"] = df["rate"].str.replace("/5", "", regex=True).astype(float)
most_ordered_type = df["listed_in(type)"].mode()[0]
```

```python
print("Most ordered type is ")
most_ordered_type
```

# **2.How many votes has each type of restaurant received from customers?**

```python
votes_per_type = df.groupby("listed_in(type)")["votes"].sum().sort_values(ascending=False)
```

```python
votes_per_type
```

# **3.What are the ratings that the majority of restaurants have received?**

```python
most_common_ratings = df["rate"].mode()[0]
```

```python
most_common_ratings
```

# **4.Zomato has observed that most couples order most of their food online. What is their average spending on each order?**


```python
avg_spending_per_order = df["approx_cost(for two people)"].mean() / 2
```

```python
avg_spending_per_order
```

# **5.Which mode (online or offline) has received the maximum rating?**

```python
online_ratings = df[df["online_order"] == "Yes"]["rate"].mean()
offline_ratings = df[df["online_order"] == "No"]["rate"].mean()
highest_rating_mode = "Online" if online_ratings > offline_ratings else "Offline"
```

```python
highest_rating_mode
```

# ***6.Which type of restaurant received more offline orders, so that Zomato can provide those customers with some good offers?***

```python
offline_orders = df[df["online_order"] == "No"]["listed_in(type)"].value_counts().idxmax()
```
